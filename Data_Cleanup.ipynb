# Data cleanup and augmentation

```python
# First we acquire the necessary imports
import cv2
import os
import os.path
# !pip install opencv-contrib-python  ## Might be necessary
import numpy as np
from sklearn.model_selection import train_test_split

np.random.seed(0)

# Esta variável serve para habilitar o redimentsionamento das imagens. Porém, as imagens originais não estão mais no repositório.
RE_SIZE_IMAGES = 0
```

### Redimenciona as imagens para o tamanho de (120, 120) pixels para facilitar a extração de features e desempenho do classificador.

```python
if RE_SIZE_IMAGES:    
    origin_dir = 'data'
    destination_dir = 'data_resized'
    output_size = (120,120)

    current_dir = destination_dir[:]

    i = 1 
    j = 0

    for pkmn in os.listdir(origin_dir):
        
        if pkmn not in os.listdir(destination_dir):
            os.mkdir(os.path.join(destination_dir, pkmn))
            
        current_dir = os.path.join(origin_dir, pkmn)
        
        print(i, pkmn)
        
        for k, img in enumerate(os.listdir(current_dir)):
            if not img.endswith(('.gif','.svg','.asp')):
                j += 1
                original_img = cv2.imread(os.path.join(current_dir, img), cv2.IMREAD_UNCHANGED)
                resized = cv2.resize(original_img, output_size)
                success = cv2.imwrite(os.path.join(destination_dir, pkmn, str(j) + '.jpg'), resized)
                
            else:
                print(pkmn,img)
        
        i += 1

    print('Done.')
```

### Com as imagens já redimencionadas, multiplicamos a mesma por 4, em diferentes posições do plano (0°, 90°, 180° e 270°) assim melhorando a extração de features do nosso classificador.

```python
origin_dir2 = 'Assets/Data_Filtered_Resized'
destination_dir2 = 'Assets/new'

def Collect_Img_Names(data_dir = origin_dir2):
    X = []
    Y = []
    for folder in os.listdir(data_dir):
        for file in os.listdir(os.path.join(data_dir,folder)):
            if not file.endswith(('.gif','.svg','.asp')):
                X.append(os.path.join(data_dir,folder,file))
                Y.append(folder)
    return X,Y

def Img_Transpose(filename, destination_dir = destination_dir2):
    for img in filename:
        if os.path.split(os.path.split(img)[0])[1] not in os.listdir(destination_dir):
            os.mkdir(os.path.join(destination_dir,os.path.split(os.path.split(img)[0])[1]))
        target = os.path.join(destination_dir, os.path.split(os.path.split(img)[0])[1], os.path.split(img)[1])
        cv2.imwrite(target,cv2.imread(img))
        
def Data_Augment(origin_dir = origin_dir2, destination_dir = destination_dir2):
    
    current_dir = destination_dir[:]
    i = 1
    
    for pkmn in os.listdir(origin_dir):
        
        if pkmn not in os.listdir(destination_dir):
            os.mkdir(os.path.join(destination_dir, pkmn))
        
        current_dir = os.path.join(origin_dir, pkmn)
        print(i, pkmn)
        j = 0
        
        for k, img in enumerate(os.listdir(current_dir)):
            
            if not img.endswith(('.gif','.svg','.asp')):
                original_img = cv2.imread(os.path.join(current_dir, img), cv2.IMREAD_UNCHANGED)
                success = cv2.imwrite(os.path.join(destination_dir, pkmn, pkmn + '_' + str(j) + '.jpg'), original_img)
                rotated = cv2.rotate(original_img, cv2.ROTATE_90_CLOCKWISE)
                success = cv2.imwrite(os.path.join(destination_dir, pkmn, pkmn + '_' + str(j+1) + '.jpg'), rotated)
                rotated = cv2.rotate(original_img, cv2.ROTATE_180)
                success = cv2.imwrite(os.path.join(destination_dir, pkmn, pkmn + '_' + str(j+2) + '.jpg'), rotated)
                rotated = cv2.rotate(original_img, cv2.ROTATE_90_COUNTERCLOCKWISE)
                success = cv2.imwrite(os.path.join(destination_dir, pkmn, pkmn + '_' + str(j+3) + '.jpg'), rotated)
                j += 4
                
            else:
                print(pkmn,img)
        i += 1
        
    print('Done.')

def Collect_Img_Names(data_dir = origin_dir2):
    X = []
    Y = []
    for folder in os.listdir(data_dir):
        for file in os.listdir(os.path.join(data_dir,folder)):
            if not file.endswith(('.gif','.svg','.asp')):
                X.append(os.path.join(data_dir,folder,file))
                Y.append(folder)
    return X,Y

def Img_Transpose(filename, destination_dir = destination_dir2):
    for img in filename:
        if os.path.split(os.path.split(img)[0])[1] not in os.listdir(destination_dir):
            os.mkdir(os.path.join(destination_dir,os.path.split(os.path.split(img)[0])[1]))
        target = os.path.join(destination_dir, os.path.split(os.path.split(img)[0])[1], os.path.split(img)[1])
        cv2.imwrite(target,cv2.imread(img))
        
def Data_Augment(origin_dir = origin_dir2, destination_dir = destination_dir2):
    
    current_dir = destination_dir[:]
    i = 1
    
    for pkmn in os.listdir(origin_dir):
        
        if pkmn not in os.listdir(destination_dir):
            os.mkdir(os.path.join(destination_dir, pkmn))
        
        current_dir = os.path.join(origin_dir, pkmn)
        print(i, pkmn)
        j = 0
        
        for k, img in enumerate(os.listdir(current_dir)):
            
            if not img.endswith(('.gif','.svg','.asp')):
                original_img = cv2.imread(os.path.join(current_dir, img), cv2.IMREAD_UNCHANGED)
                success = cv2.imwrite(os.path.join(destination_dir, pkmn, pkmn + '_' + str(j) + '.jpg'), original_img)
                rotated = cv2.rotate(original_img, cv2.ROTATE_90_CLOCKWISE)
                success = cv2.imwrite(os.path.join(destination_dir, pkmn, pkmn + '_' + str(j+1) + '.jpg'), rotated)
                rotated = cv2.rotate(original_img, cv2.ROTATE_180)
                success = cv2.imwrite(os.path.join(destination_dir, pkmn, pkmn + '_' + str(j+2) + '.jpg'), rotated)
                rotated = cv2.rotate(original_img, cv2.ROTATE_90_COUNTERCLOCKWISE)
                success = cv2.imwrite(os.path.join(destination_dir, pkmn, pkmn + '_' + str(j+3) + '.jpg'), rotated)
                j += 4
                
            else:
                print(pkmn,img)
        i += 1
        
    print('Done.')
```

```python
X,Y = Collect_Img_Names()
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.15, random_state=0)
X_train_base, X_train_super, y_train_base, y_train_super = train_test_split(X_train, y_train, test_size = 0.25, random_state = 0)

dirs = ['temp_train_base','temp_train_super', 'temp_test']
lists = [X_train_base, X_train_super, X_test]

for dir_, list_ in zip(dirs, lists):
    if 'temp' not in os.listdir("Assets"):
        os.mkdir('Assets/temp')
    if dir_ in os.listdir("Assets/temp"):
        Img_Transpose(list_, os.path.join('Assets/temp', dir_))
    else:
        os.mkdir(os.path.join('Assets/temp', dir_))
        Img_Transpose(list_, os.path.join('Assets/temp', dir_))

out_dirs = ['Train_base','Train_super', 'Test']

for dir_, out_ in zip(dirs, out_dirs):
    if 'Final' not in os.listdir("Assets"):
        os.mkdir('Assets/Final')
    if out_ in os.listdir("Assets/Final"):
        Data_Augment( os.path.join('Assets/temp', dir_), os.path.join('Assets/Final', out_))
    else:
        os.mkdir(os.path.join('Assets/Final', out_))
        Data_Augment(os.path.join('Assets/temp', dir_), os.path.join('Assets/Final', out_))
```
